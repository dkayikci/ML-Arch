```python
# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

import os
os.chdir('/content/drive/MyDrive/MetaHittite/2025enhanced_Hittite')

# Install required libraries with specific versions
!pip install numpy==1.25.2
!pip install scikit-learn==1.2.2
!pip install torch==2.3.0+cu121 torchvision==0.18.0+cu121 -f https://download.pytorch.org/whl/torch_stable.html
!pip install higher==0.2.1
!pip install evograd==0.1.2
!pip install easyfsl==1.5.0
!pip install pandas seaborn matplotlib

# Import required libraries
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torchvision.transforms as transforms
import torchvision.datasets as datasets
import torchvision.models as models
import numpy as np
import higher
import random
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from PIL import Image
from copy import deepcopy
from tqdm import tqdm
from torch.utils.data import DataLoader, Dataset, Subset, ConcatDataset, SubsetRandomSampler
from torch.optim import Adam
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.model_selection import KFold, StratifiedKFold
from torchvision.transforms import Compose, Resize, ToTensor, Normalize
from torchvision.datasets import ImageFolder
from torch.optim.lr_scheduler import ReduceLROnPlateau, CosineAnnealingLR

# Set seed for reproducibility
def set_seed(seed_value):
    """Set random seeds for reproducibility across all libraries."""
    random.seed(seed_value)
    np.random.seed(seed_value)
    torch.manual_seed(seed_value)
    if torch.cuda.is_available():
        torch.cuda.manual_seed(seed_value)
        torch.cuda.manual_seed_all(seed_value)
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False
    print(f"Seed set to {seed_value}")

# Try different seeds for better initialization
SEED = 123  # Changed from 42 to 123
set_seed(SEED)

# Define device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Enhanced ResNet18 model with additional regularization
class ResNet18HybridEnhanced(nn.Module):
    def __init__(self, num_classes=4):
        super().__init__()
        # Load pretrained ResNet18
        self.resnet = models.resnet18(pretrained=True)

        # Freeze early layers to prevent overfitting
        for name, param in self.resnet.named_parameters():
            if 'layer1' in name or 'layer2' in name:
                param.requires_grad = False

        # Get the input features of the last layer
        in_features = self.resnet.fc.in_features

        # Replace the final layer with a more complex classifier
        self.resnet.fc = nn.Sequential(
            nn.Linear(in_features, 256),
            nn.BatchNorm1d(256),
            nn.ReLU(inplace=True),
            nn.Dropout(0.4),  # Increased dropout
            nn.Linear(256, 128),
            nn.BatchNorm1d(128),
            nn.ReLU(inplace=True),
            nn.Dropout(0.3),
            nn.Linear(128, num_classes)
        )

        # Initialize the new layers
        for m in self.resnet.fc.modules():
            if isinstance(m, nn.Linear):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)
            elif isinstance(m, nn.BatchNorm1d):
                nn.init.constant_(m.weight, 1)
                nn.init.constant_(m.bias, 0)

    def forward(self, x):
        return self.resnet(x)

# Data loading and transformations
root = "/content/drive/MyDrive/MetaHittite/2025enhanced_Hittite"

# Standard transform without augmentation
standard_transform = Compose([
    Resize((224, 224)),
    ToTensor(),
    Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))
])

# Load datasets with the same transform (no augmentation)
support_dataset = ImageFolder(os.path.join(root, "support_training"), transform=standard_transform)
query_dataset = ImageFolder(os.path.join(root, "query_test"), transform=standard_transform)

# Define class names for better reporting
class_names = support_dataset.classes
print(f"Classes: {class_names}")

# Combine all data for cross-validation
all_data = ConcatDataset([support_dataset, query_dataset])

# Enhanced hyperparameters
num_epochs = 20  # Increased from 10
batch_size = 16  # Reduced from 28 for better gradient estimates
meta_lr = 0.0005      # Reduced from 0.001
inner_lr = 0.01       # Reduced from 0.02
num_inner_updates = 3  # Reduced from 5
num_meta_updates = 100  # Increased from 50
weight_decay = 0.001   # Increased from 0.0005

# Enhanced MAML training function with gradient clipping
def train_maml_enhanced(model, support_loader, query_loader, meta_lr, inner_lr,
                       num_inner_updates, num_meta_updates, weight_decay):
    """Enhanced MAML training with better optimization strategies."""
    model.to(device)
    model.train()

    # Define meta-optimizer with weight decay
    meta_optimizer = optim.Adam(model.parameters(), lr=meta_lr, weight_decay=weight_decay)

    # Learning rate scheduler for meta-optimizer
    scheduler = ReduceLROnPlateau(meta_optimizer, mode='min', factor=0.5, patience=10, verbose=True)

    # Get all data in one batch for support and query sets
    support_data, support_labels = next(iter(support_loader))
    query_data, query_labels = next(iter(query_loader))

    support_data, support_labels = support_data.to(device), support_labels.to(device)
    query_data, query_labels = query_data.to(device), query_labels.to(device)

    # Track losses for monitoring
    meta_losses = []

    # MAML training loop
    for iteration in tqdm(range(num_meta_updates), desc="MAML Training"):
        meta_optimizer.zero_grad()

        # Use class weights for balanced loss
        class_weights = torch.tensor([1.0, 1.0, 1.0, 1.0]).to(device)
        criterion = nn.CrossEntropyLoss(weight=class_weights)

        # Create a copy of the model for the inner loop
        with higher.innerloop_ctx(model, optim.SGD(model.parameters(), lr=inner_lr)) as (fast_model, fast_optim):
            # Inner loop adaptation on support set
            for inner_step in range(num_inner_updates):
                support_preds = fast_model(support_data)
                support_loss = criterion(support_preds, support_labels)
                fast_optim.step(support_loss)

            # Compute loss on query set with adapted model
            query_preds = fast_model(query_data)
            query_loss = criterion(query_preds, query_labels)

            # Add L2 regularization to the loss
            l2_lambda = 0.001
            l2_norm = sum(p.pow(2.0).sum() for p in fast_model.parameters())
            query_loss = query_loss + l2_lambda * l2_norm

            # Gradient clipping to prevent exploding gradients
            query_loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)

            meta_optimizer.step()

            # Track loss
            meta_losses.append(query_loss.item())

        # Print progress every 10 iterations
        if (iteration + 1) % 10 == 0:
            avg_loss = np.mean(meta_losses[-10:])
            print(f"Meta Iteration: {iteration+1}/{num_meta_updates}, Loss: {avg_loss:.4f}")

        # Update learning rate based on loss plateau
        if (iteration + 1) % 20 == 0:
            scheduler.step(query_loss)

    return model

# Enhanced fine-tuning function with early stopping
def fine_tune_enhanced(model, train_loader, optimizer, criterion, num_epochs, patience=5):
    """Enhanced fine-tuning with early stopping and learning rate scheduling."""
    model.train()

    # Learning rate scheduler
    scheduler = CosineAnnealingLR(optimizer, T_max=num_epochs)

    best_loss = float('inf')
    epochs_without_improvement = 0
    best_model_state = deepcopy(model.state_dict())

    for epoch in range(num_epochs):
        running_loss = 0.0
        correct = 0
        total = 0

        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)

            # Add L2 regularization
            l2_lambda = 0.001
            l2_norm = sum(p.pow(2.0).sum() for p in model.parameters())
            loss = loss + l2_lambda * l2_norm

            loss.backward()

            # Gradient clipping
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)

            optimizer.step()

            running_loss += loss.item()
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()

        # Calculate epoch statistics
        train_loss = running_loss / len(train_loader)
        train_acc = 100. * correct / total

        # Learning rate step
        scheduler.step()

        # Early stopping check
        if train_loss < best_loss:
            best_loss = train_loss
            epochs_without_improvement = 0
            best_model_state = deepcopy(model.state_dict())
        else:
            epochs_without_improvement += 1

        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {train_loss:.4f}, Acc: {train_acc:.2f}%')

        # Early stopping
        if epochs_without_improvement >= patience:
            print(f"Early stopping triggered after {epoch+1} epochs")
            model.load_state_dict(best_model_state)
            break

    return model

# Enhanced evaluation function
def evaluate_enhanced(model, loader):
    """Enhanced evaluation with additional metrics."""
    model.eval()
    all_preds = []
    all_targets = []
    all_probs = []

    with torch.no_grad():
        for inputs, targets in loader:
            inputs, targets = inputs.to(device), targets.to(device)
            outputs = model(inputs)
            probs = F.softmax(outputs, dim=1)
            _, preds = outputs.max(1)

            all_preds.extend(preds.cpu().numpy())
            all_targets.extend(targets.cpu().numpy())
            all_probs.extend(probs.cpu().numpy())

    # Calculate accuracy
    accuracy = accuracy_score(all_targets, all_preds)

    # Generate confusion matrix and classification report
    cm = confusion_matrix(all_targets, all_preds)
    cr = classification_report(all_targets, all_preds, output_dict=True,
                               target_names=class_names)

    return accuracy, cm, cr, all_targets, all_preds, np.array(all_probs)

# Enhanced visualization functions
def plot_confusion_matrix_enhanced(cm, classes, title='Confusion Matrix', cmap=plt.cm.Blues):
    """Enhanced confusion matrix visualization."""
    plt.figure(figsize=(10, 8))

    # Normalize confusion matrix
    cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    # Create heatmap with both raw counts and percentages
    sns.heatmap(cm, annot=False, fmt='d', cmap=cmap, xticklabels=classes, yticklabels=classes)

    # Add text annotations with counts and percentages
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j + 0.5, i + 0.5, f'{cm[i, j]}\n({cm_normalized[i, j]:.1%})',
                    ha='center', va='center',
                    color='white' if cm[i, j] > cm.max()/2 else 'black')

    plt.title(title)
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.tight_layout()
    plt.show()

# Enhanced cross-validation function with stratified k-fold
def run_cross_validation_enhanced(num_folds=3):
    """Enhanced k-fold cross-validation with better strategies."""
    # Use stratified k-fold for better class balance
    labels = [label for _, label in support_dataset]
    skf = StratifiedKFold(n_splits=num_folds, shuffle=True, random_state=SEED)

    # Lists to store results
    fold_train_acc = []
    fold_val_acc = []
    fold_cm = []
    fold_cr = []
    fold_models = []

    # Create query (test) dataloader
    query_loader = DataLoader(query_dataset, batch_size=len(query_dataset), shuffle=False)

    print(f"Running {num_folds}-fold cross-validation...")

    # Run stratified k-fold cross-validation
    for fold, (train_idx, val_idx) in enumerate(skf.split(range(len(support_dataset)), labels)):
        print(f"\n{'='*50}")
        print(f"FOLD {fold+1}/{num_folds}")
        print(f"{'='*50}")

        # Create data loaders for this fold
        train_sampler = SubsetRandomSampler(train_idx)
        val_sampler = SubsetRandomSampler(val_idx)

        # Use smaller batch size
        train_loader = DataLoader(support_dataset, batch_size=batch_size, sampler=train_sampler)
        val_loader = DataLoader(support_dataset, batch_size=batch_size, sampler=val_sampler)

        # Create model for this fold
        model = ResNet18HybridEnhanced(num_classes=len(class_names))
        model = model.to(device)

        # 1. Meta-Learning Phase (MAML)
        train_maml_loader = DataLoader(Subset(support_dataset, train_idx),
                                       batch_size=len(train_idx), shuffle=False)
        val_maml_loader = DataLoader(Subset(support_dataset, val_idx),
                                     batch_size=len(val_idx), shuffle=False)

        print("Starting enhanced MAML training...")
        model = train_maml_enhanced(model, train_maml_loader, val_maml_loader,
                                   meta_lr, inner_lr, num_inner_updates, num_meta_updates, weight_decay)

        # 2. Fine-tuning Phase
        print("\nStarting enhanced fine-tuning phase...")
        criterion = nn.CrossEntropyLoss()
        optimizer = Adam([
            {'params': model.resnet.fc.parameters(), 'lr': 0.001},
            {'params': model.resnet.layer4.parameters(), 'lr': 0.0001},
            {'params': model.resnet.layer3.parameters(), 'lr': 0.00001}
        ], weight_decay=weight_decay)

        model = fine_tune_enhanced(model, train_loader, optimizer, criterion, num_epochs=num_epochs)

        # 3. Evaluation Phase
        print("\nEvaluating on validation set...")
        val_acc, val_cm, val_cr, val_targets, val_preds, val_probs = evaluate_enhanced(model, val_loader)

        print("\nEvaluating on query (test) set...")
        query_acc, query_cm, query_cr, query_targets, query_preds, query_probs = evaluate_enhanced(model, query_loader)

        # Store results
        fold_train_acc.append(val_acc)  # Using validation accuracy as proxy for training
        fold_val_acc.append(query_acc)
        fold_cm.append(query_cm)
        fold_cr.append(query_cr)
        fold_models.append(deepcopy(model.state_dict()))

        # Print results
        print(f"\nValidation Accuracy: {val_acc:.4f}")
        print(f"Query (Test) Accuracy: {query_acc:.4f}")

        print("\nValidation Confusion Matrix:")
        print(val_cm)

        print("\nQuery Confusion Matrix:")
        print(query_cm)

        print("\nValidation Classification Report:")
        print(classification_report(val_targets, val_preds, target_names=class_names))

        print("\nQuery Classification Report:")
        print(classification_report(query_targets, query_preds, target_names=class_names))

        # Visualize results
        plot_confusion_matrix_enhanced(val_cm, class_names, f"Fold {fold+1} - Validation Confusion Matrix")
        plot_confusion_matrix_enhanced(query_cm, class_names, f"Fold {fold+1} - Query Confusion Matrix")

    # Calculate average metrics
    avg_train_acc = np.mean(fold_train_acc)
    avg_val_acc = np.mean(fold_val_acc)

    # Find best fold
    best_fold_idx = np.argmax(fold_val_acc)

    print(f"\n{'='*50}")
    print(f"CROSS-VALIDATION RESULTS")
    print(f"{'='*50}")
    print(f"Average Training Accuracy: {avg_train_acc:.4f}")
    print(f"Average Validation Accuracy: {avg_val_acc:.4f}")
    print(f"Best Fold: {best_fold_idx+1} with Accuracy: {fold_val_acc[best_fold_idx]:.4f}")
    print(f"Standard Deviation: {np.std(fold_val_acc):.4f}")

    # Load the best model for final evaluation
    best_model = ResNet18HybridEnhanced(num_classes=len(class_names))
    best_model.load_state_dict(fold_models[best_fold_idx])
    best_model = best_model.to(device)

    # Evaluate best model on the query set
    query_loader = DataLoader(query_dataset, batch_size=6, shuffle=False)
    query_acc, query_cm, query_cr, _, _, _ = evaluate_enhanced(best_model, query_loader)

    print(f"\n{'='*50}")
    print(f"BEST MODEL EVALUATION")
    print(f"{'='*50}")
    print(f"Best Model Query Accuracy: {query_acc:.4f}")
    print("\nBest Model Query Confusion Matrix:")
    print(query_cm)

    # Visualize the best model's confusion matrix
    plot_confusion_matrix_enhanced(query_cm, class_names, "Best Model Query Confusion Matrix")

    # Save the best model
    model_name = f'enhanced_hybrid_model_seed{SEED}_noaug.pth'
    torch.save(best_model.state_dict(), model_name)
    print(f"\nBest model saved as '{model_name}'")

    return {
        'avg_train_acc': avg_train_acc,
        'avg_val_acc': avg_val_acc,
        'best_fold_idx': best_fold_idx,
        'best_fold_acc': fold_val_acc[best_fold_idx],
        'best_model': best_model,
        'fold_results': {
            'train_acc': fold_train_acc,
            'val_acc': fold_val_acc,
            'confusion_matrices': fold_cm,
            'classification_reports': fold_cr
        }
    }

# Run the enhanced cross-validation experiment
results = run_cross_validation_enhanced(num_folds=3)

# Additional analysis and visualization
print("\nClass-specific performance for best model:")
best_cr = results['fold_results']['classification_reports'][results['best_fold_idx']]

# Create summary table
summary_data = []
for cls in class_names:
    summary_data.append({
        'Class': cls,
        'Precision': best_cr[cls]['precision'],
        'Recall': best_cr[cls]['recall'],
        'F1-Score': best_cr[cls]['f1-score']
    })

summary_df = pd.DataFrame(summary_data)
print(summary_df)

# Plot performance across folds
plt.figure(figsize=(10, 6))
folds = range(1, len(results['fold_results']['val_acc']) + 1)
plt.plot(folds, results['fold_results']['train_acc'], 'o-', label='Validation Accuracy', linewidth=2)
plt.plot(folds, results['fold_results']['val_acc'], 's-', label='Test Accuracy', linewidth=2)
plt.xlabel('Fold')
plt.ylabel('Accuracy')
plt.title('Model Performance Across Folds')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Create box plot for class-wise performance
plt.figure(figsize=(12, 8))
class_performance = []
for fold_cr in results['fold_results']['classification_reports']:
    for cls in class_names:
        class_performance.append({
            'Class': cls,
            'F1-Score': fold_cr[cls]['f1-score'],
            'Metric': 'F1-Score'
        })
        class_performance.append({
            'Class': cls,
            'Precision': fold_cr[cls]['precision'],
            'Metric': 'Precision'
        })
        class_performance.append({
            'Class': cls,
            'Recall': fold_cr[cls]['recall'],
            'Metric': 'Recall'
        })

class_df = pd.DataFrame(class_performance)
sns.boxplot(x='Class', y='F1-Score', data=class_df[class_df['Metric'] == 'F1-Score'])
plt.title('F1-Score Distribution by Class Across Folds')
plt.ylim(0, 1)
plt.tight_layout()
plt.show()

print("\nEnhanced Hybrid Model Training Complete!")
print(f"Final Test Accuracy: {results['best_fold_acc']:.4f}")
print(f"Standard Deviation: {np.std(results['fold_results']['val_acc']):.4f}")
```
