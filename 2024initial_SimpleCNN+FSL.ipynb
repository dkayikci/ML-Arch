```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torchvision import transforms
from torch.utils.data import Dataset, DataLoader, SubsetRandomSampler
from torch.optim.lr_scheduler import StepLR
from PIL import Image
import os
import random
import numpy as np
from sklearn.model_selection import KFold
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, accuracy_score, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

# Set the device to GPU if available, otherwise use CPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Define image transformations
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # Resize images to 224x224
    transforms.RandomHorizontalFlip(),  # Random horizontal flip for data augmentation
    transforms.RandomVerticalFlip(),  # Random vertical flip for data augmentation
    transforms.RandomRotation(20),  # Random rotation for data augmentation
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),  # Color jitter for data augmentation
    transforms.ToTensor(),  # Convert image to tensor
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  # Normalize image
])

# Custom dataset class for few-shot learning
class FewShotDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.root_dir = root_dir
        self.transform = transform
        self.classes = [d for d in os.listdir(root_dir) if os.path.isdir(os.path.join(root_dir, d))]
        self.class_to_idx = {cls_name: i for i, cls_name in enumerate(self.classes)}
        self.images = self._load_images()

    def _load_images(self):
        images = []
        for cls in self.classes:
            class_dir = os.path.join(self.root_dir, cls)
            for img_name in os.listdir(class_dir):
                images.append((os.path.join(class_dir, img_name), self.class_to_idx[cls]))
        return images

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img_path, label = self.images[idx]
        image = Image.open(img_path).convert('RGB')
        if self.transform:
            image = self.transform(image)
        return image, label

# Define the CNN model
class CNNModel(nn.Module):
    def __init__(self, num_classes):
        super(CNNModel, self).__init__()
        self.conv1 = nn.Conv2d(3, 64, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm2d(64)
        self.conv2 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm2d(128)
        self.conv3 = nn.Conv2d(128, 256, kernel_size=3, padding=1)
        self.bn3 = nn.BatchNorm2d(256)
        self.pool = nn.MaxPool2d(2, 2)
        self.adaptive_pool = nn.AdaptiveAvgPool2d((7, 7))
        self.dropout = nn.Dropout(0.5)
        self.fc1 = nn.Linear(256 * 7 * 7, 1024)
        self.fc2 = nn.Linear(1024, num_classes)

    def forward(self, x):
        x = self.pool(F.relu(self.bn1(self.conv1(x))))
        x = self.pool(F.relu(self.bn2(self.conv2(x))))
        x = self.pool(F.relu(self.bn3(self.conv3(x))))
        x = self.adaptive_pool(x)
        x = x.view(-1, 256 * 7 * 7)
        x = self.dropout(F.relu(self.fc1(x)))
        x = self.fc2(x)
        return x

# Function to train the model for one epoch
def train_epoch(model, optimizer, dataloader, device):
    model.train()
    total_loss = 0
    total_accuracy = 0
    all_true = []
    all_pred = []

    for batch_images, batch_labels in dataloader:
        batch_images, batch_labels = batch_images.to(device), batch_labels.to(device)

        optimizer.zero_grad()

        outputs = model(batch_images)
        loss = F.cross_entropy(outputs, batch_labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        predictions = torch.argmax(outputs, dim=1)
        total_accuracy += (predictions == batch_labels).float().mean().item()

        all_true.extend(batch_labels.cpu().numpy())
        all_pred.extend(predictions.cpu().numpy())

    return total_loss / len(dataloader), total_accuracy / len(dataloader), all_true, all_pred

# Function to evaluate the model
def evaluate(model, dataloader, device):
    model.eval()
    total_accuracy = 0
    all_true = []
    all_pred = []

    with torch.no_grad():
        for batch_images, batch_labels in dataloader:
            batch_images, batch_labels = batch_images.to(device), batch_labels.to(device)

            outputs = model(batch_images)
            predictions = torch.argmax(outputs, dim=1)
            total_accuracy += (predictions == batch_labels).float().mean().item()

            all_true.extend(batch_labels.cpu().numpy())
            all_pred.extend(predictions.cpu().numpy())

    return total_accuracy / len(dataloader), all_true, all_pred

# Function to plot and print confusion matrix
def plot_confusion_matrix(y_true, y_pred, classes, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
    plt.title(title)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()

    print(f"\n{title}")
    print(cm)

# Main function
def main():
    # Set paths for train and test data
    train_data_root = '/content/drive/MyDrive/MetaHittite/stelea1500x1000pxsL2/FSL/supportL2_112'
    test_data_root = '/content/drive/MyDrive/MetaHittite/stelea1500x1000pxsL2/FSL/queryL2_24'

    # Create datasets
    train_dataset = FewShotDataset(train_data_root, transform=transform)
    test_dataset = FewShotDataset(test_data_root, transform=transform)

    # Set hyperparameters
    batch_size = 16  #.........
    num_epochs = 20
    n_folds = 2

    # Initialize K-Fold cross-validation
    kf = KFold(n_splits=n_folds, shuffle=True, random_state=42)

    fold_results = []
    test_predictions = []

    for fold, (train_idx, val_idx) in enumerate(kf.split(train_dataset.images)):
        print(f"Fold {fold + 1}/{n_folds}")

        # Create data loaders for this fold
        train_sampler = SubsetRandomSampler(train_idx)
        val_sampler = SubsetRandomSampler(val_idx)

        train_loader = DataLoader(train_dataset, batch_size=batch_size, sampler=train_sampler)
        val_loader = DataLoader(train_dataset, batch_size=batch_size, sampler=val_sampler)
        test_loader = DataLoader(test_dataset, batch_size=batch_size)

        # Initialize model, optimizer, and scheduler
        model = CNNModel(num_classes=len(train_dataset.classes)).to(device)
        optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-4)
        scheduler = StepLR(optimizer, step_size=10, gamma=0.5)

        best_val_acc = 0
        train_losses = []
        train_accuracies = []
        val_accuracies = []

        # Training loop
        for epoch in range(num_epochs):
            train_loss, train_acc, _, _ = train_epoch(model, optimizer, train_loader, device)
            val_acc, _, _ = evaluate(model, val_loader, device)

            train_losses.append(train_loss)
            train_accuracies.append(train_acc)
            val_accuracies.append(val_acc)

            print(f"Epoch {epoch+1}/{num_epochs}")
            print(f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f}, Val Acc: {val_acc:.4f}")

            if val_acc > best_val_acc:
                best_val_acc = val_acc
                torch.save(model.state_dict(), f"best_model_fold_{fold+1}.pth")

            scheduler.step()

        # Load best model for this fold
        model.load_state_dict(torch.load(f"best_model_fold_{fold+1}.pth"))

        # Evaluate on train set
        train_acc, train_true, train_pred = evaluate(model, train_loader, device)

        # Evaluate on test set
        test_acc, test_true, test_pred = evaluate(model, test_loader, device)

        fold_results.append({
            'train_losses': train_losses,
            'train_accuracies': train_accuracies,
            'val_accuracies': val_accuracies,
            'train_accuracy': train_acc,
            'test_accuracy': test_acc,
            'train_true': train_true,
            'train_pred': train_pred,
            'test_true': test_true,
            'test_pred': test_pred
        })

        print(f"Fold {fold+1} Train Accuracy: {train_acc:.4f}")
        print(f"Fold {fold+1} Test Accuracy: {test_acc:.4f}")

        # Plot training progress
        plt.figure(figsize=(15, 5))
        plt.subplot(1, 3, 1)
        plt.plot(train_losses)
        plt.title(f'Fold {fold+1} Training Loss')
        plt.subplot(1, 3, 2)
        plt.plot(train_accuracies)
        plt.title(f'Fold {fold+1} Training Accuracy')
        plt.subplot(1, 3, 3)
        plt.plot(val_accuracies)
        plt.title(f'Fold {fold+1} Validation Accuracy')
        plt.tight_layout()
        plt.show()

        # Plot and print confusion matrices
        plot_confusion_matrix(train_true, train_pred, train_dataset.classes, f'Fold {fold+1} Train Confusion Matrix')
        plot_confusion_matrix(test_true, test_pred, test_dataset.classes, f'Fold {fold+1} Test Confusion Matrix')

    # Aggregate results across folds
    avg_train_acc = np.mean([res['train_accuracy'] for res in fold_results])
    std_train_acc = np.std([res['train_accuracy'] for res in fold_results])
    avg_test_acc = np.mean([res['test_accuracy'] for res in fold_results])
    std_test_acc = np.std([res['test_accuracy'] for res in fold_results])

    print(f"\nAverage Train Accuracy: {avg_train_acc:.4f} ± {std_train_acc:.4f}")
    print(f"Average Test Accuracy: {avg_test_acc:.4f} ± {std_test_acc:.4f}")

    # Compute overall metrics
    all_train_true = []
    all_train_pred = []
    all_test_true = []
    all_test_pred = []
    for res in fold_results:
        all_train_true.extend(res['train_true'])
        all_train_pred.extend(res['train_pred'])
        all_test_true.extend(res['test_true'])
        all_test_pred.extend(res['test_pred'])

    # Calculate and print final performance metrics
    print("\nFinal Performance Report:")
    print("\nTrain Set Metrics:")
    print(classification_report(all_train_true, all_train_pred, target_names=train_dataset.classes))

    print("\nTest Set Metrics:")
    print(classification_report(all_test_true, all_test_pred, target_names=test_dataset.classes))

    # Plot overall confusion matrices
    plot_confusion_matrix(all_train_true, all_train_pred, train_dataset.classes, 'Overall Train Confusion Matrix')
    plot_confusion_matrix(all_test_true, all_test_pred, test_dataset.classes, 'Overall Test Confusion Matrix')

if __name__ == '__main__':
    main()
```
