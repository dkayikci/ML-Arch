```python
# Mount Google Drive and setup
from google.colab import drive
drive.mount('/content/drive')

import os
os.chdir('/content/drive/MyDrive/MetaHittite/2025enhanced_Hittite')

# Install required libraries
!pip install numpy==1.25.2
!pip install torch==2.3.0+cu121 torchvision==0.18.0+cu121 -f https://download.pytorch.org/whl/torch_stable.html
!pip install scikit-learn==1.2.2
!pip install higher==0.2.1
!pip install evograd==0.1.2
!pip install easyfsl==1.5.0
!pip install seaborn pandas matplotlib tqdm pillow

# Import libraries
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torchvision import transforms
from torch.utils.data import Dataset, DataLoader, SubsetRandomSampler, WeightedRandomSampler
from torch.optim.lr_scheduler import StepLR, ReduceLROnPlateau, CosineAnnealingLR
from PIL import Image
import os
import random
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.utils.class_weight import compute_class_weight
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from collections import Counter

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Set random seed for reproducibility
def set_seed(seed):
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    np.random.seed(seed)
    random.seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed(123)  # Changed seed for potentially better initialization

# Dataset class
class HittiteDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.root_dir = root_dir
        self.transform = transform
        self.classes = sorted([d for d in os.listdir(root_dir) if os.path.isdir(os.path.join(root_dir, d))])
        self.class_to_idx = {cls_name: i for i, cls_name in enumerate(self.classes)}
        self.images = self._load_images()

        # Count samples per class
        self.class_counts = Counter([label for _, label in self.images])
        print(f"Class distribution: {dict(self.class_counts)}")

    def _load_images(self):
        images = []
        for cls in self.classes:
            class_dir = os.path.join(self.root_dir, cls)
            for img_name in os.listdir(class_dir):
                if img_name.lower().endswith(('.jpg', '.jpeg', '.png')):
                    images.append((os.path.join(class_dir, img_name), self.class_to_idx[cls]))
        return images

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img_path, label = self.images[idx]
        image = Image.open(img_path).convert('RGB')

        if self.transform:
            image = self.transform(image)

        return image, label

# Simplified CNN architecture to prevent overfitting
class SimplerCNN(nn.Module):
    def __init__(self, num_classes):
        super(SimplerCNN, self).__init__()
        # Fewer layers and parameters
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm2d(32)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm2d(64)
        self.conv3 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        self.bn3 = nn.BatchNorm2d(128)

        self.pool = nn.MaxPool2d(2, 2)
        self.avgpool = nn.AdaptiveAvgPool2d((7, 7))

        # Strong dropout
        self.dropout = nn.Dropout(0.5)

        # Fewer FC parameters
        self.fc1 = nn.Linear(128 * 7 * 7, 256)
        self.fc2 = nn.Linear(256, num_classes)

        # Initialize weights
        self._initialize_weights()

    def _initialize_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                nn.init.xavier_normal_(m.weight)
            elif isinstance(m, nn.BatchNorm2d):
                nn.init.constant_(m.weight, 1)
                nn.init.constant_(m.bias, 0)
            elif isinstance(m, nn.Linear):
                nn.init.xavier_normal_(m.weight)
                nn.init.constant_(m.bias, 0)

    def forward(self, x):
        x = self.pool(F.relu(self.bn1(self.conv1(x))))
        x = self.pool(F.relu(self.bn2(self.conv2(x))))
        x = self.pool(F.relu(self.bn3(self.conv3(x))))

        x = self.avgpool(x)
        x = x.view(x.size(0), -1)

        x = self.dropout(F.relu(self.fc1(x)))
        x = self.fc2(x)

        return x

# Focal Loss for handling class imbalance
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2):
        super(FocalLoss, self).__init__()
        self.alpha = alpha
        self.gamma = gamma

    def forward(self, inputs, targets):
        ce_loss = F.cross_entropy(inputs, targets, reduction='none')
        pt = torch.exp(-ce_loss)
        focal_loss = (1 - pt) ** self.gamma * ce_loss

        if self.alpha is not None:
            alpha_t = self.alpha[targets]
            focal_loss = alpha_t * focal_loss

        return focal_loss.mean()

# Training function with label smoothing
def train_epoch(model, optimizer, dataloader, device, criterion):
    model.train()
    total_loss = 0
    correct = 0
    total = 0

    for batch_images, batch_labels in dataloader:
        batch_images, batch_labels = batch_images.to(device), batch_labels.to(device)

        optimizer.zero_grad()
        outputs = model(batch_images)

        # Label smoothing
        smoothing = 0.1
        n_classes = outputs.size(1)
        targets = torch.zeros_like(outputs)
        targets.fill_(smoothing / (n_classes - 1))
        targets.scatter_(1, batch_labels.unsqueeze(1), 1 - smoothing)

        # Use KL divergence loss for label smoothing
        loss = F.kl_div(F.log_softmax(outputs, dim=1), targets, reduction='batchmean')

        # Add L2 regularization
        l2_lambda = 0.01
        l2_norm = sum(p.pow(2.0).sum() for p in model.parameters())
        loss = loss + l2_lambda * l2_norm

        loss.backward()

        # Gradient clipping
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)

        optimizer.step()

        total_loss += loss.item()
        _, predicted = outputs.max(1)
        total += batch_labels.size(0)
        correct += predicted.eq(batch_labels).sum().item()

    return total_loss / len(dataloader), correct / total

# Evaluation function
def evaluate(model, dataloader, device):
    model.eval()
    correct = 0
    total = 0
    predictions = []
    labels = []

    with torch.no_grad():
        for batch_images, batch_labels in dataloader:
            batch_images, batch_labels = batch_images.to(device), batch_labels.to(device)
            outputs = model(batch_images)
            _, predicted = outputs.max(1)

            total += batch_labels.size(0)
            correct += predicted.eq(batch_labels).sum().item()

            predictions.extend(predicted.cpu().numpy())
            labels.extend(batch_labels.cpu().numpy())

    return correct / total, predictions, labels

# Main training function
def main():
    # Data paths
    train_data_root = '/content/drive/MyDrive/MetaHittite/2025enhanced_Hittite/support_training'
    test_data_root = '/content/drive/MyDrive/MetaHittite/2025enhanced_Hittite/query_test'

    # Simple transform without augmentation
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])

    # Create datasets
    train_dataset = HittiteDataset(train_data_root, transform=transform)
    test_dataset = HittiteDataset(test_data_root, transform=transform)

    # Calculate class weights for imbalanced classes
    labels = [label for _, label in train_dataset.images]
    class_weights = compute_class_weight(
        'balanced',
        classes=np.unique(labels),
        y=labels
    )
    class_weights = torch.tensor(class_weights, dtype=torch.float32).to(device)
    print(f"Class weights: {class_weights}")

    # Hyperparameters
    batch_size = 8  # Smaller batch size
    num_epochs = 100  # Fewer epochs
    patience = 10  # Shorter patience

    # Stratified K-fold cross-validation
    skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=123)
    fold_results = []

    # Extract labels for stratification
    all_labels = np.array([label for _, label in train_dataset.images])

    for fold, (train_idx, val_idx) in enumerate(skf.split(np.zeros(len(all_labels)), all_labels)):
        print(f"\n{'='*50}")
        print(f"FOLD {fold + 1}/3")
        print(f"{'='*50}")

        # Create samplers for balanced batching
        train_labels = all_labels[train_idx]
        class_sample_counts = np.bincount(train_labels)
        weights = 1.0 / torch.tensor(class_sample_counts, dtype=torch.float32)
        train_targets = torch.tensor(train_labels)
        samples_weights = weights[train_targets]
        train_sampler = WeightedRandomSampler(samples_weights, len(samples_weights))

        val_sampler = SubsetRandomSampler(val_idx)

        # Create dataloaders
        train_loader = DataLoader(train_dataset, batch_size=batch_size, sampler=train_sampler)
        val_loader = DataLoader(train_dataset, batch_size=batch_size, sampler=val_sampler)
        test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

        # Initialize model
        model = SimplerCNN(num_classes=len(train_dataset.classes)).to(device)

        # Focal loss for class imbalance
        criterion = FocalLoss(alpha=class_weights, gamma=2)

        # Optimizer with different learning rates
        optimizer = optim.SGD([
            {'params': model.conv1.parameters(), 'lr': 0.001},
            {'params': model.conv2.parameters(), 'lr': 0.001},
            {'params': model.conv3.parameters(), 'lr': 0.001},
            {'params': model.fc1.parameters(), 'lr': 0.01},
            {'params': model.fc2.parameters(), 'lr': 0.01}
        ], momentum=0.9, weight_decay=0.01)

        # Learning rate scheduler
        scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.5, patience=5, verbose=True)

        # Training variables
        best_val_acc = 0
        no_improve_epochs = 0
        train_losses = []
        train_accuracies = []
        val_accuracies = []

        # Training loop
        for epoch in range(num_epochs):
            # Train
            train_loss, train_acc = train_epoch(model, optimizer, train_loader, device, criterion)

            # Validate
            val_acc, _, _ = evaluate(model, val_loader, device)

            # Store metrics
            train_losses.append(train_loss)
            train_accuracies.append(train_acc)
            val_accuracies.append(val_acc)

            # Learning rate scheduling
            scheduler.step(val_acc)

            # Early stopping
            if val_acc > best_val_acc:
                best_val_acc = val_acc
                torch.save(model.state_dict(), f'best_model_fold_{fold+1}.pth')
                no_improve_epochs = 0
            else:
                no_improve_epochs += 1

            # Print progress
            if epoch % 5 == 0:
                print(f"Epoch {epoch+1}/{num_epochs}")
                print(f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f}")
                print(f"Val Acc: {val_acc:.4f}, Best Val Acc: {best_val_acc:.4f}")

            # Early stopping
            if no_improve_epochs >= patience:
                print(f"Early stopping at epoch {epoch+1}")
                break

        # Load best model
        model.load_state_dict(torch.load(f'best_model_fold_{fold+1}.pth'))

        # Final evaluation
        train_acc_final, train_pred, train_true = evaluate(model, train_loader, device)
        test_acc, test_pred, test_true = evaluate(model, test_loader, device)

        print(f"\nFold {fold+1} Final Results:")
        print(f"Train Accuracy: {train_acc_final:.4f}")
        print(f"Test Accuracy: {test_acc:.4f}")

        # Plot confusion matrix
        cm = confusion_matrix(test_true, test_pred)
        plt.figure(figsize=(10, 8))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                    xticklabels=train_dataset.classes,
                    yticklabels=train_dataset.classes)
        plt.title(f'Fold {fold+1} Test Confusion Matrix')
        plt.xlabel('Predicted')
        plt.ylabel('True')
        plt.show()

        # Print classification report
        print("\nClassification Report:")
        print(classification_report(test_true, test_pred, target_names=train_dataset.classes))

        # Store results
        fold_results.append({
            'train_accuracy': train_acc_final,
            'test_accuracy': test_acc,
            'test_true': test_true,
            'test_pred': test_pred
        })

    # Calculate average metrics
    avg_train_acc = np.mean([res['train_accuracy'] for res in fold_results])
    avg_test_acc = np.mean([res['test_accuracy'] for res in fold_results])
    std_train_acc = np.std([res['train_accuracy'] for res in fold_results])
    std_test_acc = np.std([res['test_accuracy'] for res in fold_results])

    print(f"\n{'='*50}")
    print(f"FINAL RESULTS")
    print(f"{'='*50}")
    print(f"Average Train Accuracy: {avg_train_acc:.4f} ± {std_train_acc:.4f}")
    print(f"Average Test Accuracy: {avg_test_acc:.4f} ± {std_test_acc:.4f}")

    # Aggregate all predictions
    all_test_true = []
    all_test_pred = []
    for res in fold_results:
        all_test_true.extend(res['test_true'])
        all_test_pred.extend(res['test_pred'])

    # Final classification report
    print("\nFinal Performance Report (Aggregated over Folds):")
    print(classification_report(all_test_true, all_test_pred,
                              target_names=train_dataset.classes))

    # Plot overall confusion matrix
    cm_overall = confusion_matrix(all_test_true, all_test_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm_overall, annot=True, fmt='d', cmap='Blues',
                xticklabels=train_dataset.classes,
                yticklabels=train_dataset.classes)
    plt.title('Overall Test Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()

if __name__ == '__main__':
    main()
```
