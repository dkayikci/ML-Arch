```python
# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Install specific versions of required libraries
!pip install numpy==1.25.2 scikit-learn==1.2.2


import os
os.chdir('/content/drive/MyDrive/MetaHittite/2025enhanced_Hittite')

# Import necessary libraries
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, f1_score, confusion_matrix

# Print directory structure for reference
def print_directory_structure(path, level=0):
    print('    ' * level + os.path.basename(path) + '/')
    if level < 2:  # Limit depth to avoid excessive output
        for item in os.listdir(path):
            item_path = os.path.join(path, item)
            if os.path.isdir(item_path):
                print_directory_structure(item_path, level + 1)
            else:
                print('    ' * (level + 1) + item)

folder_path = "/content/drive/MyDrive/MetaHittite/2025enhanced_Hittite"
print("Directory structure:")
print_directory_structure(folder_path)

# Define paths for support (training) and query (test) datasets
support_path = os.path.join(folder_path, "support_training")
query_path = os.path.join(folder_path, "query_test")

# Function to load and preprocess images
def load_images_from_folder(folder, target_size=(224, 224)):
    images = []
    labels = []
    class_names = []

    # Get class folders
    class_folders = [f for f in os.listdir(folder) if os.path.isdir(os.path.join(folder, f))]
    class_folders.sort()  # Ensure consistent order

    # Map class names to numeric labels
    for i, class_name in enumerate(class_folders):
        class_names.append(class_name)
        class_path = os.path.join(folder, class_name)

        for img_name in os.listdir(class_path):
            img_path = os.path.join(class_path, img_name)

            # Read and resize image
            img = cv2.imread(img_path)
            if img is not None:
                img = cv2.resize(img, target_size)
                # Flatten the image
                img_flat = img.flatten()
                images.append(img_flat)
                labels.append(i)

    return np.array(images), np.array(labels), class_names

# Load training and testing data
print("Loading support (training) data...")
X_train, y_train, class_names = load_images_from_folder(support_path)
print("Loading query (testing) data...")
X_test, y_test, _ = load_images_from_folder(query_path)

print(f"Training data shape: {X_train.shape}")
print(f"Testing data shape: {X_test.shape}")
print(f"Class names: {class_names}")

# Standardize the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Function to train, evaluate and visualize results for a model
def train_and_evaluate_model(model, X_train, y_train, X_test, y_test):
    # Train the model
    model.fit(X_train, y_train)

    # Make predictions on train and test sets
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    # Calculate metrics for train set
    train_accuracy = accuracy_score(y_train, y_train_pred)
    train_precision = precision_score(y_train, y_train_pred, average='weighted')
    train_f1 = f1_score(y_train, y_train_pred, average='weighted')

    # Calculate metrics for test set
    test_accuracy = accuracy_score(y_test, y_test_pred)
    test_precision = precision_score(y_test, y_test_pred, average='weighted')
    test_f1 = f1_score(y_test, y_test_pred, average='weighted')

    # Perform cross-validation
    cv_scores = cross_val_score(model, X_train, y_train, cv=3)

    # Create confusion matrices
    train_cm = confusion_matrix(y_train, y_train_pred)
    test_cm = confusion_matrix(y_test, y_test_pred)

    return {
        'train': {
            'accuracy': train_accuracy,
            'precision': train_precision,
            'f1': train_f1,
            'cm': train_cm,
            'predictions': y_train_pred
        },
        'test': {
            'accuracy': test_accuracy,
            'precision': test_precision,
            'f1': test_f1,
            'cm': test_cm,
            'predictions': y_test_pred
        },
        'cv_scores': cv_scores
    }

# Function to print confusion matrix in a readable format
def print_confusion_matrix(cm, classes, title):
    print(f"{title} Confusion Matrix:")
    print("    " + " ".join(f"{c:>5}" for c in classes))
    for i, row in enumerate(cm):
        print(f"{classes[i]:>3} " + " ".join(f"{cell:>5}" for cell in row))
    print()

# Function to plot confusion matrix
def plot_confusion_matrix(cm, classes, title):
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
    plt.title(title)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.tight_layout()
    plt.show()

# Function to evaluate and display results
def evaluate_and_print_results(model_name, results, classes):
    print(f"\n{model_name} Results:")

    print("Training Set Performance:")
    print(f"Accuracy: {results['train']['accuracy']:.4f}")
    print(f"Precision: {results['train']['precision']:.4f}")
    print(f"F1 Score: {results['train']['f1']:.4f}")
    print_confusion_matrix(results['train']['cm'], classes, "Training")
    plot_confusion_matrix(results['train']['cm'], classes, f"{model_name} - Training Confusion Matrix")

    print("Test Set Performance:")
    print(f"Accuracy: {results['test']['accuracy']:.4f}")
    print(f"Precision: {results['test']['precision']:.4f}")
    print(f"F1 Score: {results['test']['f1']:.4f}")
    print_confusion_matrix(results['test']['cm'], classes, "Test")
    plot_confusion_matrix(results['test']['cm'], classes, f"{model_name} - Test Confusion Matrix")

    print(f"Cross-validation Scores: {results['cv_scores']}")
    print(f"Average CV Score: {np.mean(results['cv_scores']):.4f}")

    # Return metrics for summary table
    return {
        'train_acc': results['train']['accuracy'],
        'test_acc': results['test']['accuracy'],
        'cv_score': np.mean(results['cv_scores'])
    }

# Define the models to evaluate
models = {
    'SVM': SVC(kernel='rbf', random_state=42),
    'KNN': KNeighborsClassifier(n_neighbors=5),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Logistic Regression': LogisticRegression(multi_class='ovr', random_state=42, max_iter=1000),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Naive Bayes': GaussianNB()
}

# Train and evaluate each model
results_summary = {}
for model_name, model in models.items():
    print(f"\nEvaluating {model_name}...")
    results = train_and_evaluate_model(model, X_train_scaled, y_train, X_test_scaled, y_test)
    metrics = evaluate_and_print_results(model_name, results, class_names)
    results_summary[model_name] = metrics

# Create a summary table
print("\n\nSummary of Model Performance:")
print("-" * 80)
print(f"{'Model':<20} {'Training Accuracy':<20} {'Test Accuracy':<20} {'CV Score':<20}")
print("-" * 80)
for model_name, metrics in results_summary.items():
    print(f"{model_name:<20} {metrics['train_acc']:.4f}{'':<14} {metrics['test_acc']:.4f}{'':<14} {metrics['cv_score']:.4f}")
print("-" * 80)

# Visualize model comparisons
plt.figure(figsize=(14, 8))

# Set up bar positions
models_list = list(results_summary.keys())
x = np.arange(len(models_list))
width = 0.25

# Create bars
plt.bar(x - width, [results_summary[m]['train_acc'] for m in models_list], width, label='Training Accuracy')
plt.bar(x, [results_summary[m]['test_acc'] for m in models_list], width, label='Test Accuracy')
plt.bar(x + width, [results_summary[m]['cv_score'] for m in models_list], width, label='CV Score')

# Add labels and legend
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Model Performance Comparison')
plt.xticks(x, models_list, rotation=45)
plt.ylim(0, 1.1)
plt.legend()
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add horizontal line at 0.625 for human expert reference
plt.axhline(y=0.625, color='r', linestyle='-', alpha=0.5)
plt.text(len(models_list)-1, 0.65, 'Human Expert (62.5%)', color='r')

plt.show()

# Print final conclusions
print("\nConclusions:")
print("------------")
print("1. All conventional ML models show clear signs of overfitting (high training accuracy, low test accuracy)")
print("2. Logistic Regression performs best on test data with 50% accuracy")
print("3. No conventional model outperforms the human expert benchmark of 62.5%")
print("4. These results support the use of more advanced approaches (transfer learning, meta-learning)")
print("5. The spatial information loss from flattening images severely impacts model performance")
```
