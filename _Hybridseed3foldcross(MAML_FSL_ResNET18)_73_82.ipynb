```python
import os
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.transforms as transforms
import torchvision.models as models
import numpy as np
import higher
import random
import matplotlib.pyplot as plt
from PIL import Image
from torchvision.models import resnet18
from torch.utils.data import DataLoader, Dataset, Subset
from torch.optim import Adam
from sklearn.metrics import confusion_matrix, classification_report
from torchvision.transforms import Compose, Resize, ToTensor, Normalize
from torchvision.datasets import ImageFolder
from sklearn.model_selection import StratifiedKFold
import seaborn as sns
import pandas as pd

# Set seed for reproducibility
def set_seed(seed):
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    np.random.seed(seed)
    random.seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed(42)

class ResNet18Modified(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        self.resnet = resnet18(pretrained=True)
        self.resnet.fc = nn.Linear(512, num_classes)

    def forward(self, x):
        return self.resnet(x)

# Data loading and preprocessing
root = "/content/drive/MyDrive/MetaHittite/stelea1500x1000pxsL2/FSL"
transform = Compose([Resize((224, 224)), ToTensor(), Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))])

support_dataset = ImageFolder(os.path.join(root, "supportL2_112"), transform=transform)
query_dataset = ImageFolder(os.path.join(root, "queryL2_24"), transform=transform)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

def maml_train(model, support_dataloader, query_dataloader, meta_lr=0.01, inner_lr=0.02, num_inner_updates=6, num_meta_updates=70, weight_decay=0.0006):
    meta_optimizer = optim.Adam(model.parameters(), lr=meta_lr, weight_decay=weight_decay)
    inner_optimizer = optim.SGD(model.parameters(), lr=inner_lr)

    for meta_iteration in range(num_meta_updates):
        support_data, support_labels = next(iter(support_dataloader))
        query_data, query_labels = next(iter(query_dataloader))

        support_data, support_labels = support_data.to(device), support_labels.to(device)
        query_data, query_labels = query_data.to(device), query_labels.to(device)

        model.train()
        meta_optimizer.zero_grad()

        with higher.innerloop_ctx(model, inner_optimizer) as (fmodel, diffopt):
            for _ in range(num_inner_updates):
                support_preds = fmodel(support_data)
                inner_loss = nn.CrossEntropyLoss()(support_preds, support_labels)
                diffopt.step(inner_loss)

            query_preds = fmodel(query_data)
            outer_loss = nn.CrossEntropyLoss()(query_preds, query_labels)
            outer_loss.backward()
            meta_optimizer.step()

        if meta_iteration % 5 == 0:
            print(f"Meta-iteration {meta_iteration}: Outer Loss: {outer_loss.item()}")

def train_model(model, dataloader, criterion, optimizer, num_epochs=25):
    for epoch in range(num_epochs):
        for images, labels in dataloader:
            images, labels = images.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')

def evaluate_model(model, dataloader):
    model.eval()
    all_labels = []
    all_preds = []

    with torch.no_grad():
        for images, labels in dataloader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)
            all_labels.extend(labels.cpu().numpy())
            all_preds.extend(predicted.cpu().numpy())

    return all_labels, all_preds

def plot_confusion_matrix(cm, title):
    plt.figure(figsize=(10, 7))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(title)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()

def plot_classification_report(class_report, title):
    metrics = ['precision', 'recall', 'f1-score']
    class_names = list(class_report.keys())[:-3]
    data = {metric: [class_report[class_name][metric] for class_name in class_names] for metric in metrics}
    df = pd.DataFrame(data, index=class_names)

    df.plot(kind='bar', figsize=(10, 8))
    plt.title(title)
    plt.ylabel('Score')
    plt.xlabel('Classes')
    plt.ylim(0, 1)
    plt.show()

# 3-fold cross-validation
def cross_validate(support_dataset, query_dataset, n_splits=3):
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)

    fold_results = []

    for fold, (train_idx, val_idx) in enumerate(skf.split(support_dataset.samples, support_dataset.targets)):
        print(f"Fold {fold + 1}/{n_splits}")

        # Create data loaders for this fold
        train_subset = Subset(support_dataset, train_idx)
        val_subset = Subset(support_dataset, val_idx)

        train_loader = DataLoader(train_subset, batch_size=28, shuffle=True)
        val_loader = DataLoader(val_subset, batch_size=28, shuffle=False)
        query_loader = DataLoader(query_dataset, batch_size=6, shuffle=False)

        # Initialize and train the model
        model = ResNet18Modified(num_classes=4).to(device)

        # MAML training
        maml_train(model, train_loader, DataLoader(query_dataset, batch_size=6, shuffle=True))

        # Fine-tuning
        criterion = nn.CrossEntropyLoss()
        optimizer = Adam(model.parameters(), lr=0.002)
        train_model(model, train_loader, criterion, optimizer, num_epochs=50)

        # Evaluate
        val_labels, val_preds = evaluate_model(model, val_loader)
        query_labels, query_preds = evaluate_model(model, query_loader)

        # Compute metrics
        val_cm = confusion_matrix(val_labels, val_preds)
        val_report = classification_report(val_labels, val_preds, output_dict=True)
        query_cm = confusion_matrix(query_labels, query_preds)
        query_report = classification_report(query_labels, query_preds, output_dict=True)

        fold_results.append({
            'val_cm': val_cm,
            'val_report': val_report,
            'query_cm': query_cm,
            'query_report': query_report
        })

        # Print fold results
        print(f"Validation Set Performance:")
        print(classification_report(val_labels, val_preds))
        print(f"Query Set Performance:")
        print(classification_report(query_labels, query_preds))
        print("=" * 50)

    return fold_results

# Run cross-validation
cv_results = cross_validate(support_dataset, query_dataset)

# Compute average results
avg_val_cm = np.mean([fold['val_cm'] for fold in cv_results], axis=0).astype(int)
avg_query_cm = np.mean([fold['query_cm'] for fold in cv_results], axis=0).astype(int)

avg_val_report = {}
avg_query_report = {}

for key in cv_results[0]['val_report'].keys():
    if key not in ['accuracy', 'macro avg', 'weighted avg']:
        avg_val_report[key] = {metric: np.mean([fold['val_report'][key][metric] for fold in cv_results])
                               for metric in ['precision', 'recall', 'f1-score']}
        avg_query_report[key] = {metric: np.mean([fold['query_report'][key][metric] for fold in cv_results])
                                 for metric in ['precision', 'recall', 'f1-score']}

# Print average results
print("Average Validation Confusion Matrix:")
print(avg_val_cm)
print("\nAverage Query Confusion Matrix:")
print(avg_query_cm)

print("\nAverage Validation Classification Report:")
print(pd.DataFrame(avg_val_report).T)
print("\nAverage Query Classification Report:")
print(pd.DataFrame(avg_query_report).T)

# Visualizations
plot_confusion_matrix(avg_val_cm, "Average Validation Confusion Matrix")
plot_confusion_matrix(avg_query_cm, "Average Query Confusion Matrix")
plot_classification_report(avg_val_report, "Average Validation Classification Report")
plot_classification_report(avg_query_report, "Average Query Classification Report")

# Calculate and print average accuracies
avg_val_accuracy = np.mean([fold['val_report']['accuracy'] for fold in cv_results])
avg_query_accuracy = np.mean([fold['query_report']['accuracy'] for fold in cv_results])

print(f"\nAverage Validation Accuracy: {avg_val_accuracy:.4f}")
print(f"Average Query Accuracy: {avg_query_accuracy:.4f}")
```
